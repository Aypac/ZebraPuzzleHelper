```python
from ZebraPuzzle import ZebraPuzzle
```

```python
puz = ZebraPuzzle({'house colors': ['red', 'green', 'white', 'yellow', 'blue'],
                   'house position': ['left', 'center-left', 'center', 'center-right', 'right'],
                   'smoke': ['Pall mall', 'Blends', 'Bluemasters', 'Prince', 'Dunhill'],
                   'nationality': ['Norwegian', 'British', 'Swedish', 'Danish', 'German'],
                   'pet': ['dog', 'fish', 'bird', 'horse', 'cat'],
                   'drink': ['water', 'milk', 'beer', 'tea', 'coffee'],
                   })
```

```python
#1
puz.add_assoc_name('british', 'red', True)

#2
puz.add_assoc_name('dog', 'swedish', True)

#3
puz.add_assoc_name('tea', 'danish', True)

#4
puz.add_assoc_name('white', 'left', False)
puz.add_assoc_name('green', 'right', False)

#5
puz.add_assoc_name('coffee', 'green', True)

#6
puz.add_assoc_name('bird', 'pall mall', True)

#7
puz.add_assoc_name('dunhill', 'yellow', True)

#8
puz.add_assoc_name('milk', 'center', True)

#9
puz.add_assoc_name('norwegian', 'left', True)

#15
puz.add_assoc_name('blends', 'cat', False)

#11
puz.add_assoc_name('horse', 'yellow', False)

#11
puz.add_assoc_name('blends', 'cat', False)

#12
puz.add_assoc_name('beer', 'bluemasters', True)

#13
puz.add_assoc_name('german', 'prince', True)

#14
puz.add_assoc_name('norwegian', 'blue', False)
#14+9
puz.add_assoc_name('blue', 'center-left', True)

#15
puz.add_assoc_name('blends', 'water', False)

#Info left in: 4, 10, 11, 15
puz.display_matrix()
```

```python
puz.iterate_logic()
puz.display_matrix()
```

```python
#4
puz.add_assoc_name('white', 'center-right', False)
puz.add_assoc_name('green', 'left', False)

#Info left in: 4, 10, 11, 15

puz.iterate_logic()
puz.display_matrix()
```

```python
#4
puz.add_assoc_name('white', 'right', True)

#Info left in: 10, 11, 15

puz.iterate_logic()
puz.display_matrix()
```

```python
#15
puz.add_assoc_name('blends', 'center-left', True)

#Info left in: 10, 11

puz.iterate_logic()
puz.display_matrix()
```

```python
#15
puz.add_assoc_name('cat', 'center-left', False)
puz.add_assoc_name('cat', 'center-right', False)
puz.add_assoc_name('cat', 'right', False)

#11
puz.add_assoc_name('horse', 'center-left', True)

#Info left in: 10

puz.iterate_logic()
puz.display_matrix()
```

So we see, that the German, Prince-smoking person, coffee-drinker living in the green-house at the center-right owns the fish.
